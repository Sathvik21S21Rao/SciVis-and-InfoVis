```python
#Calculating the average of the stringency index for each country for each month 

df['Date'] = pd.to_datetime(df['Date'])
df['Month'] = df['Date'].dt.month
df['Year'] = df['Date'].dt.year

df['Month'] = df['Month'].astype(str)
df['Year'] = df['Year'].astype(str)

df['Month_Year'] = df['Month'] + '-' + df['Year']

df['Month_Year'] = pd.to_datetime(df['Month_Year'], format='%m-%Y')

df['Month_Year'] = df['Month_Year'].dt.strftime('%m-%Y')

df
```

```python
df.drop(['E1_Income support', 'E1_Flag','E2_Debt/contract relief', 'E3_Fiscal measures','E4_International support', 'Month', 'Year'], axis=1, inplace=True)
```

```python
#Calculating the average for each of the containment measures for each country for each month

df.head()
```

```python
# List of columns to average
columns_to_average = [
    'C1_School closing', 'C2_Workplace closing', 'C3_Cancel public events',
    'C4_Restrictions on gatherings', 'C5_Close public transport',
    'C6_Stay at home requirements', 'C7_Restrictions on internal movement',
    'C8_International travel controls'
]

# Group by 'CountryName' and 'Month_Year' and calculate the mean for the specified columns
average_values = df.groupby(['CountryName', 'Month_Year'])[columns_to_average].mean().reset_index()

# Display the resulting DataFrame
average_values
```

```python
# Step 1: Convert 'Month_Year' to datetime format
average_values['Month_Year'] = pd.to_datetime(average_values['Month_Year'], format='%m-%Y')

# Step 2: Sort the values by 'CountryName' and the properly formatted 'Month_Year'
average_values = average_values.sort_values(by=['CountryName', 'Month_Year'])

# (Optional) If you want to convert 'Month_Year' back to a string after sorting, you can do so:
average_values['Month_Year'] = average_values['Month_Year'].dt.strftime('%m-%Y')

# Display the sorted DataFrame
average_values
```

```python
# Sum the values of C1 to C8 columns for each row
average_values['Total'] = average_values[columns_to_average].sum(axis=1)

# Sort the DataFrame by the 'Total' column in descending order
sorted_average_values = average_values.sort_values(by='Total', ascending=False)

# Drop the 'Total' column if you don't need it anymore
sorted_average_values = sorted_average_values.drop(columns=['Total'])

# Display the sorted DataFrame
sorted_average_values.CountryName.unique()[:10]
```

```python
cleaned_df = sorted_average_values[sorted_average_values['CountryName'].isin(['Philippines', 'Chile', 'Honduras', 'Kazakhstan',
       'Kyrgyz Republic', 'Libya', 'Georgia', 'El Salvador', 'Argentina',
       'Serbia'])]
```

```python
cleaned_df
```

```python
# Step 1: Convert 'Month_Year' to datetime format
cleaned_df['Month_Year'] = pd.to_datetime(cleaned_df['Month_Year'], format='%m-%Y')

# Step 2: Sort the values by 'CountryName' and the properly formatted 'Month_Year'
cleaned_df = cleaned_df.sort_values(by=['CountryName', 'Month_Year'])

# (Optional) If you want to convert 'Month_Year' back to a string after sorting, you can do so:
cleaned_df['Month_Year'] = cleaned_df['Month_Year'].dt.strftime('%m-%Y')

# Display the sorted DataFrame
cleaned_df.reset_index(drop=True, inplace=True)

cleaned_df.CountryName.unique()
```
